# Figures

```python
import os
import numpy as np
import matplotlib.pyplot as plt

FIG_PATH = "./figs/"
```

```python
from scipy.stats import norm

num = 500
x = np.linspace(-5, 10, num)
p1 = norm.pdf(x, loc=0, scale=1)
p2 = norm.pdf(x, loc=5, scale=2)

plt.xticks([], [])
plt.yticks([], [])

plt.plot(p1)
plt.plot(p2)

plt.savefig(os.path.join(FIG_PATH, "ChoosingHP.svg"), bbox_inches="tight")
plt.show()
```
